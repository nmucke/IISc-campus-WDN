```python
# import libraries
import pandas as pd


from ML_for_WDN.time_series_utils import (
    prepare_time_series,
)
```

```python
sensors = [1, 2, 3, 4, 6, 8, 9, 11, 13]

df_no_leak = pd.read_excel(
    '../data/raw_data/no_leak.xlsx', 
    skiprows=0,
    sheet_name=None,
    
    header=0,
)

df_no_leak = prepare_time_series(df_no_leak, sensors)

df_no_leak.to_csv('../data/processed_data/no_leak.csv', index=False)
```

```python
df_leak = pd.read_excel(
    '../data/raw_data/leak_loc_1.xlsx', 
    skiprows=0,
    sheet_name=None,
    header=0,
)

df_leak = prepare_time_series(df_leak, sensors)

df_leak.to_csv('../data/processed_data/leak_loc_1.csv', index=False)
```
